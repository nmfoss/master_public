```python
import pandas as pd
```

```python
def recode_topics(topic):
    recoded = topic
    if len(topic) == 3:
        recoded = topic[:1]
    if len(topic) == 4:
        recoded = topic[:2]
    return recoded
```

```python
df_merged = pd.read_pickle('../pandas/merged_df.pkl')
```

```python
df_merged
```

```python
df_merged.columns = ['raw', 'lemma_delivered', 'label', 'date', 'pub', 'coder_id']
df_merged['agg_label'] = df_merged['label'].fillna(-1).astype(int).astype(str).apply(recode_topics).astype(int)
df_merged['label'] = df_merged['label'].fillna(-1).astype(int)
df_merged['pub'] = df_merged['pub'].astype('category')
df_merged['date'] = pd.to_datetime(df_merged['date'])
df_merged['coder_id'].fillna(0, inplace=True)
df_merged['coder_id'] = df_merged['coder_id'].astype('category')
```

```python
df_merged
```

```python
df_merged['raw_len'] = df_merged['raw'].astype(str).apply(len)
df_merged['raw_word_count'] = df_merged['raw'].apply(lambda x: len(str(x).split()))
```
